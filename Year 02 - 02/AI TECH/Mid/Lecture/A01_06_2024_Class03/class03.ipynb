## **Class 03: Support Vector Machine (SVM)**

```python
import cv2
from matplotlib import pyplot as plt
```

### **General**
- Just another algorithm like Linear Regression
- Usage: Linear Reg. - approximate, Logistic Reg. - classification
- **Feature**: the thing that is used to seperate data, it can be used directly to process the data. (input: image, feature: length, color, etc.)
- AI: automatic system doing something for you

### **Evolution of AI**
- **First age (AI)**: just simple system automatically doing things for us, calculate if-else with human set ***threshold***
- **Second age (ML)**: calculate threshold automatically using mathematical equation and ***dataset***
- **Last age (DL)**: automatically choose the ***feature*** (noramlly need an expert to determine) -> but require a lot of resources (data, computation)
- NOTE: *DL ⊆ ML ⊆ AI*

### **SVM Example**
- 2 features case

```python
img = cv2.imread('img01.jpg')
plt.imshow(img)
```

- for two feature, the algorithm generate the best fit line with the largeast margin between the data
- the margin will make the model more **robust** to noise 
- featres will be treat as vector, using **dot product**
> **Dot Product**
> - find projection
> - find how 2 vector are similar to each other
> - *v ⋅ u = |v||u|cosθ*



```python
img = cv2.imread('img02.jpg')
plt.figure(figsize=(10, 10))
plt.imshow(img)
```

- only **3 closet** data influence the position of the seperating line (AKA. *hyperplane.*) - (more than 3 will conflict one another)
- principle is to find the hyperplane that creates the largest margin.

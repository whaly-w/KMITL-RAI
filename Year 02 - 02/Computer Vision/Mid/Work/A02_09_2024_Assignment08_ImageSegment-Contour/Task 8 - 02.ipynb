```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load Image
img = cv2.imread('69.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
```

```python
# Crop Image
img_cropped = img[50:185,20:205]

# Threshold
_, img_thr = cv2.threshold(img_cropped, 80, 255, cv2.THRESH_BINARY_INV)

# Display
plt.imshow(img_thr, cmap= 'gray')
```

```python
# Get Contour
contours, _ = cv2.findContours(img_thr, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Prepare image to draw
img_contour = cv2.cvtColor(img_cropped, cv2.COLOR_GRAY2BGR)

# Draw contour
cv2.drawContours(img_contour, contours, -1, (255, 0, 0), 1)

# Draw bounding rectangle
for contour in contours:
    # Get bounding rectangle
    x,y,w,h = cv2.boundingRect(contour)
    
    # Check the contour size
    if cv2.contourArea(contour) > 200:
        color = (0, 255, 0)
    else:
        color = (0, 0, 255)
        
    # Draw regtangle
    img_contour = cv2.rectangle(img_contour, (x, y), (x+w, y+h), color, 2)

# Display image
plt.imshow(img_contour)
```

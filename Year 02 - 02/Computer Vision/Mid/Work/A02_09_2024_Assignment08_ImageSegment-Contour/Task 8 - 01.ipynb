```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load Image
img = cv2.imread('coins.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
```

```python
# Pre process

# Thresholding
_, img_thr = cv2.threshold(img, 45, 50, cv2.THRESH_BINARY)

# Closing 
img_closed = cv2.morphologyEx(img_thr, cv2.MORPH_CLOSE,
                              np.ones((3,3), np.uint8),
                              iterations= 2)

# Display
plt.figure(figsize= (11,11))
plt.subplot(1, 2, 1)
plt.imshow(img_thr, cmap= 'gray')

plt.subplot(1, 2, 2)
plt.imshow(img_closed, cmap='gray')
```

```python
### -- Euclidean
# Apply distance transform
dist = cv2.distanceTransform(img_closed, cv2.DIST_L2, 5)
cv2.normalize(dist, dist, 0, 1.0, cv2.NORM_MINMAX)

# Display
plt.title('Euclidean')
plt.imshow(dist, cmap='gray')
```

```python
### -- City Block
# Apply distance transform
dist = cv2.distanceTransform(img_closed, cv2.DIST_L1, 5)
cv2.normalize(dist, dist, 0, 1.0, cv2.NORM_MINMAX)

# Display
plt.title('City Block')
plt.imshow(dist, cmap='gray')
```

```python
### -- Chess Board
# Apply distance transform
dist = cv2.distanceTransform(img_closed, cv2.DIST_C, 5)
cv2.normalize(dist, dist, 0, 1.0, cv2.NORM_MINMAX)

# Display
plt.title('Chess Board')
plt.imshow(dist, cmap='gray')
```

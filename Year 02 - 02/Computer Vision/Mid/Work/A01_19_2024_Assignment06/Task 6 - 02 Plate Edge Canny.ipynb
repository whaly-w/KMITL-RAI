```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load Image
img = cv2.imread('69.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')

def imgDisplay(localImg):
    plt.figure(figsize= (11, 11))
    
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap= 'gray')
    
    plt.subplot(1, 2, 2)
    plt.imshow(localImg, cmap= 'gray')
    
    plt.show()
```

```python
# Apply blur

# Define kernel
Gx = cv2.getGaussianKernel(5, 1.4)
Gy = np.transpose(Gx)
kernel = Gx * Gy

# Apply kernel
img_GausKer = cv2.filter2D(img, -1, kernel)

# Display
imgDisplay(img_GausKer)
```

```python
threshold = [(50, 80), (80, 100), (100, 200), (150, 200)]

img_canny = []
for (mn, mx) in threshold:
    edge = cv2.Canny(img, mn, mx)
    img_canny.append((edge, f'mx:{mx}, mn:{mn}'))
    

plt.figure(figsize= (11, 11))
for i, img in enumerate(img_canny):
    plt.subplot(2, 2, i+1)
    plt.imshow(img[0], cmap= 'gray')
    plt.title(img[1])
plt.show()
```

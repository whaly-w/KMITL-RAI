```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load Image
img = cv2.imread('coins.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')

def imgDisplay(localImg):
    plt.figure(figsize= (11, 11))
    
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap= 'gray')
    
    plt.subplot(1, 2, 2)
    plt.imshow(localImg, cmap= 'gray')
    
    plt.show()
```

```python
# Task 5.1.1 Average Kernel

# Define kernel
kernel = np.ones((15, 15)) / (15**2)

# Apply kernel
img_AvgKer = cv2.filter2D(img, -1, kernel)

# Display
imgDisplay(img_AvgKer)
```

```python
# Task 5.1.2 Gaussian Kernel

# Define kernel
Gx = cv2.getGaussianKernel(15, 0.7)
Gy = np.transpose(Gx)
kernel = Gx * Gy

# Apply kernel
img_GausKer = cv2.filter2D(img, 0, kernel)

# Display
imgDisplay(img_GausKer)
```

```python
# Task 5.1.3 Sharp Kernel

# Define kernel
kernel = np.float32([[1, 4, 6, 4, 1],
                 [4, 16, 24, 16, 4],
                 [6, 24, -476, 24, 6],
                 [4, 16, 24, 16, 4],
                 [1, 4, 6, 4, 1]]) / (-256)
print(kernel)

# Apply kernel
img_ShpKer = cv2.filter2D(img, -1, kernel)

# Display
imgDisplay(img_ShpKer)
```

#### ***Lecture 03***  | Image Transformation

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load Image
img = cv2.imread('flower.jpg')

y, x, color = img.shape
img_size = (x, y)

plt.imshow(img)
```

```python
### --- Scaling ---

# Create matrix:    [scale   0    0 ]
#                   [  0   scale  0 ]
M1 = np.float32([[0.5, 0, 0],
                [0, 0.5, 0]])

img_M1 = cv2.warpAffine(img, M1, img_size)
plt.imshow(img_M1)
```

```python
### --- Transition ---

# Create matrix:    [1 0 x]
#                   [0 1 y]
M1 = np.float32([[1, 0, 120], [0, 1, 100]])

img_M1 = cv2.warpAffine(img, M1, img_size)

plt.imshow(img_M1)
```

```python
### --- Rotation ---
# Use "getRotationMatrix2D()" to generate transformation matrix
# > params : cord, angle, size
#   - cord: rotation point
#   - angle: rotation angle
#   - size: size factor
M1 = cv2.getRotationMatrix2D((x/2, y/2), -90, 0.5)
print(M1)

img_M1 = cv2.warpAffine(img, M1, img_size)

plt.imshow(img_M1)
```

```python
### --- Shear ---

# Create matrix:   [   1   shr_x   0 ]
#                  [ shr_y   1     0 ] 
M1 = np.float32([[1, 0.6, 0],
                [0, 1, 0]])

img_M1 = cv2.warpAffine(img, M1, img_size)

plt.imshow(img_M1)
```

```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Load image
pic = cv2.imread('colorobject.png')
rgb = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(pic, cv2.COLOR_BGR2HSV)


fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(pic)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(hsv)
```

```python
# Task01: Yellow
upper_yel = (50, 255, 255)
lower_yel = (30, 50, 50)

# Create mask for yellow region
mask_yel = cv2.inRange(hsv, lower_yel, upper_yel)
mask_yel = cv2.cvtColor(mask_yel, cv2.COLOR_GRAY2BGR)

# Mask yellow region
yel = cv2.bitwise_and(rgb, mask_yel)

# Display
plt.imshow(yel)
```

```python
# Task02: Blue
upper_blue = (140, 255, 255)
lower_blue = (120, 50, 50)

# Create mask for yellow region
mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_blue = cv2.cvtColor(mask_blue, cv2.COLOR_GRAY2BGR)

# Mask yellow region
blue = cv2.bitwise_and(rgb, mask_blue)

# Display
plt.imshow(blue)
```

```python
# Task03: Blue & Yello

# Combine blue n' yellow mask
mask_BY = mask_blue + mask_yel

# Mask blue n' yellow region
BY = cv2.bitwise_and(rgb, mask_BY)

# Display
plt.imshow(BY)
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Task 3.2

# Load Image
img = cv2.imread('messi.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
sz = img.shape
img_sz = (sz[1], sz[0])


# Crop image
img_crop = cv2.getRectSubPix(img, (60, 60), (362, 314))

# Resize resolution
img_res = cv2.resize(img_crop, (300, 300))


# Display
plt.figure(figsize=(12, 10))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(img_crop)
plt.title('Cropped')

plt.subplot(1, 3, 3)
plt.imshow(img_res)
plt.title('Resized')

plt.show()
```

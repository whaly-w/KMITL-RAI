```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load Image
img = cv2.imread('coins.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')

def imgDisplay(localImg):
    plt.figure(figsize= (11, 11))
    
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap= 'gray')
    
    plt.subplot(1, 2, 2)
    plt.imshow(localImg, cmap= 'gray')
    
    plt.show()
```

```python
# Apply threshold
ret, img_thresh = cv2.threshold(img, 45, 255, cv2.THRESH_BINARY)

plt.imshow(img_thresh, cmap= 'gray')
```

```python
# Reduce noise
img_res = cv2.morphologyEx(img_thresh, cv2.MORPH_CLOSE, np.ones((3,3)), iterations= 2)

plt.imshow(img_res, cmap= 'gray')
```

```python
# Create border
ker1 = np.ones((3,3), np.uint8)
img_dilated = cv2.dilate(img_res, ker1, iterations= 3)
border = img_res - img_dilated

# Recolor
res = cv2.merge([img_res, img_dilated, img_res])

# Display
plt.figure(figsize= (11, 11))
plt.subplot(1, 2, 1)
plt.imshow(border, cmap= 'gray')
plt.subplot(1, 2, 2)
plt.imshow(res)
```

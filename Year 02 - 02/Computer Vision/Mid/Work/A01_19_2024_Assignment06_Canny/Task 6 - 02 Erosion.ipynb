```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Load Image
img = cv2.imread('j.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')

def imgDisplay(localImg):
    plt.figure(figsize= (11, 11))
    
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap= 'gray')
    
    plt.subplot(1, 2, 2)
    plt.imshow(localImg, cmap= 'gray')
    
    plt.show()
```

```python
# Erode image
kernel = np.ones((5,5), np.uint8)
img_eroded = cv2.erode(img, kernel, iterations= 1)

imgDisplay(img_eroded)
```

```python
## Soluiton 1
# Convert img to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

# Create mask
mask = img - img_eroded

# Mask a specific region to blue
img_rgb[mask > 0] = [0, 0, 255]

# Display
imgDisplay(img_rgb)
```

```python
## Solution 2

# Create edge
edge = img - img_eroded

# Adjust the intensity of each color
# | low -> approach black
# | high -> appraoch the color
r = np.uint8(img_eroded * 0)
g = np.uint8(img_eroded * 1)
b = np.uint8(img_eroded * 0)

# Merge to create rgb-colored image
img_rgb = cv2.merge([r, g, b])

# Adjust color for edge
r = np.uint8(edge * 0)
g = np.uint8(edge * 0)
b = np.uint8(edge * 1)
edge_rgb = cv2.merge([r, g, b])

# Combine the edge with the image
img_result = img_rgb + edge_rgb

# Display
imgDisplay(img_result)
```

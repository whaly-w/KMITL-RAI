```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load Image
img = cv2.imread('Source\cameraman.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')

def imgDisplay(localImg1, localImg2):
    plt.figure(figsize= (11, 11))
    
    plt.subplot(1, 2, 1)
    plt.imshow(localImg1, cmap= 'gray')
    
    plt.subplot(1, 2, 2)
    plt.imshow(localImg2, cmap= 'gray')
    
    plt.show()
```

```python
ret, res1 = cv2.threshold(img_gray, 100, 255, cv2.THRESH_TRUNC)
ret, res2 = cv2.threshold(img_gray, 100, 255, cv2.THRESH_TOZERO)
ret, res3 = cv2.threshold(img_gray, 100, 255, cv2.THRESH_TOZERO_INV)

plt.figure(figsize= (11, 11))

plt.subplot(1, 3, 1)
plt.imshow(res1, cmap='gray')

plt.subplot(1, 3, 2)
plt.imshow(res2, cmap='gray')

plt.subplot(1, 3, 3)
plt.imshow(res3, cmap='gray')

plt.show()
```

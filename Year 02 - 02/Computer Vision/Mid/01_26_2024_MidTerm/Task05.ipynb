```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load Image
img = cv2.cvtColor(cv2.imread('Source/flower.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')

row, col, c = img.shape
img_size = (col, row)
print(img_size)


def imgDisplay(localImg1, localImg2):
    plt.figure(figsize= (11, 11))
    
    plt.subplot(1, 2, 1)
    plt.imshow(localImg1, cmap= 'gray')
    
    plt.subplot(1, 2, 2)
    plt.imshow(localImg2, cmap= 'gray')
    
    plt.show()
```

```python
M1 = np.float32([[0.5, 0, 480],
                [0, 0.5, 180]])

img_M1 = cv2.warpAffine(img, M1, img_size)
plt.imshow(img_M1)
```

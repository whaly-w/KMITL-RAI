```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load Image
img = cv2.imread('Source/j.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')

def imgDisplay(localImg1, localImg2):
    plt.figure(figsize= (11, 11))
    
    plt.subplot(1, 2, 1)
    plt.imshow(localImg1, cmap= 'gray')
    
    plt.subplot(1, 2, 2)
    plt.imshow(localImg2, cmap= 'gray')
    
    plt.show()
```

```python
# Dilate image
kernel = np.ones((7,7), np.uint8)
img_dilated = cv2.dilate(img, kernel, iterations= 1)

imgDisplay(img, img_dilated)
```

```python
# Get edge
edge = img_dilated - img

# Get background
bg = cv2.bitwise_not(img_dilated)

imgDisplay(edge, bg)
```

```python
# Color each matrix

# J
r = np.uint8(img * 1)
g = np.uint8(img * 1)
b = np.uint8(img * 0)
img_rgb = cv2.merge([r, g, b])

# Edge
r = np.uint8(edge * 1)
g = np.uint8(edge * 0)
b = np.uint8(edge * 0)
edge_rgb = cv2.merge([r, g, b])

# Background
r = np.uint8(bg * 0)
g = np.uint8(bg * 1)
b = np.uint8(bg * 0)
bg_rgb = cv2.merge([r, g, b])

# Combine
img_result = img_rgb + edge_rgb + bg_rgb


# Display
imgDisplay(img, img_result)
```
